```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm trích xuất vector FFT
def extract_fft_vector(frame, fft_size):
    return np.fft.fft(frame, fft_size)

# Số chiều FFT
N_FFT = 512

# Tạo một ví dụ về khung tín hiệu
frame_signal = np.sin(2 * np.pi * 50 * np.linspace(0, 1, N_FFT)) + 0.5 * np.random.normal(size=N_FFT)

# Trích xuất vector FFT
fft_result = extract_fft_vector(frame_signal, N_FFT)

# Vẽ biểu đồ tần số
plt.figure(figsize=(20,10))
plt.plot(np.abs(fft_result))
plt.title('Biểu đồ tần số của khung tín hiệu')
plt.xlabel('Tần số')
plt.ylabel('Biên độ')
plt.show()
```
