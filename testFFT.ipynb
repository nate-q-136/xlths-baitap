```python
import numpy as np
import matplotlib.pyplot as plt

# Hm trch xut vector FFT
def extract_fft_vector(frame, fft_size):
    return np.fft.fft(frame, fft_size)

# S chiu FFT
N_FFT = 512

# To mt v d v khung tn hiu
frame_signal = np.sin(2 * np.pi * 50 * np.linspace(0, 1, N_FFT)) + 0.5 * np.random.normal(size=N_FFT)

# Trch xut vector FFT
fft_result = extract_fft_vector(frame_signal, N_FFT)

# V biu  tn s
plt.figure(figsize=(20,10))
plt.plot(np.abs(fft_result))
plt.title('Biu  tn s ca khung tn hiu')
plt.xlabel('Tn s')
plt.ylabel('Bin ')
plt.show()
```
