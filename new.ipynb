```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


# create vector n of 41 elements (time base of signal)
n = np.arange(0, 41, 1)

plt.figure(figsize=(20,10))
# create sine functions with different frequencies
w = 0
x = np.cos(w * n)  # w0=0
plt.subplot(5, 2, 1)
plt.stem(n, x)
plt.title('w=0')

w = np.pi/8
x = np.cos(w * n)  # w1=pi/4
plt.subplot(5, 2, 3)
plt.stem(n, x)
plt.title('w=pi/8')

w = np.pi/4
x = np.cos(w * n)  # w0=0
plt.subplot(5, 2, 5)
plt.stem(n, x)
plt.title('w=pi/4')

w = np.pi/2
x = np.cos(w * n)  # w1=pi/4
plt.subplot(5, 2, 7)
plt.stem(n, x)
plt.title('w=pi/2')

w = np.pi
x = np.cos(w * n)  # w1=pi/4
plt.subplot(5, 2, 9)
plt.stem(n, x)
plt.title('w=pi')

# ----------
w = 0
x = np.cos((2*np.pi - w) * n)  # w=2pi
plt.subplot(5, 2, 2)
plt.stem(n, x)
plt.title('w=2pi-0')

w = np.pi/8
x = np.cos((2*np.pi - w) * n)  # w=2pi-pi/8
plt.subplot(5, 2, 4)
plt.stem(n, x)
plt.title('w=2pi-pi/8')

w = np.pi/4
x = np.cos((2*np.pi - w) * n)  # w=2pi-pi/4
plt.subplot(5, 2, 6)
plt.stem(n, x)
plt.title('w=2pi-pi/4')

w = np.pi/2
x = np.cos((2*np.pi - w) * n)  # w=2pi-pi/2
plt.subplot(5, 2, 8)
plt.stem(n, x)
plt.title('w=2pi-pi/2')

w = np.pi
x = np.cos((2*np.pi - w) * n)  # w=2pi-pi
plt.subplot(5, 2, 10)
plt.stem(n, x)
plt.title('w=2pi-pi')

plt.show()

# Figure 2
x1 = np.cos(2*np.pi*31/10*n)  # ts chuan hoa w=31/10 -> tuan hoan voi N=10
plt.figure(2)
plt.subplot(2, 1, 1)
plt.stem(n, x1)
plt.title('w=31/10')

x2 = np.cos(2*np.pi*np.pi*n)  # ts chuan hoa w=pi -> ko tuan hoan
plt.subplot(2, 1, 2)
plt.stem(n, x2)
plt.title('w=pi')

plt.show()
```
